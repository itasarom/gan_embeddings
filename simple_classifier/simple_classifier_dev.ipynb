```python
import numpy as np
import tensorflow as tf
import pandas as pd

import matplotlib.pyplot as plt

from collections import namedtuple
import seaborn as sns
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
import data_processing as dp
import average_classifier as ac
import model_utilities as mu 
import evaluation as ev
```

```python
vocab, all_labels, sents, labels = dp.load_problem(lang='ma')
```

```python
cls = ac.AverageClassifier(vocab, n_topics=len(all_labels))
```

```python
batch_sampler = mu.BatchSampler(sents=sents, labels=labels, vocab=vocab, all_labels=all_labels, batch_size=64)
```

```python
trainer = mu.Trainer(cls)
```

```python
trainer.restore_latest("./checkpoints/")
```

```python
trainer.train(batch_sampler=batch_sampler, n_epochs=200, max_iterations=10**5, save_path="./checkpoints/checkpoint.ckpt", save_every=10)
```

```python
x, mask, y = batch_sampler.get_test()
```
